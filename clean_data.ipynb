# Merge price and coffee data

```python
arabica_df.head()
```

```python
arabica_df['year'].value_counts()
```

```python
i = 0 
while i < 1312: 
    for row in arabica_df['Country of Origin']:
        if row == 'Ethiopia':
            arabica_df.set_value(str(i), 'Price', price_df.at[row, arabica_df.at[str(i), 'year']])
        i += 1
```

```python
excluded_countries = ['United States', 'United States (Hawaii)', 'China', 'Taiwan', 'Tanzania, United Republic Of', 'Japan', 'United States (Puerto Rico)', 'Laos', 'Myanmar', 
'Mauritius', 'Cote d?Ivoire']
i = 0 
while i < 1312: 
    for row in arabica_df['Country of Origin']:
        print(row, i)
        if row not in excluded_countries and i != 1197:
            if arabica_df.at[str(i), 'year'] == '13':
                arabica_df.set_value(str(i), 'Price', price_df.at[row, '20'+arabica_df.at[str(i), 'year']])
            elif arabica_df.at[str(i), 'year'] == '17':
                arabica_df.set_value(str(i), 'Price', price_df.at[row, '20'+arabica_df.at[str(i), 'year']])
            else: 
                arabica_df.set_value(str(i), 'Price', price_df.at[row, arabica_df.at[str(i), 'year']])
        i += 1
```
